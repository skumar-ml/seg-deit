```python
import cv2
import skimage
from skimage.data import astronaut
from skimage.util import img_as_float
from skimage.segmentation import slic, mark_boundaries, felzenszwalb
import numpy as np
import matplotlib.pyplot as plt



# img = img_as_float(astronaut()[::2, ::2])
simple = cv2.imread('simple_car.png')
simple = cv2.cvtColor(simple, cv2.COLOR_BGR2RGB)

complex = cv2.imread('complex_car.jpeg')
complex = cv2.cvtColor(complex, cv2.COLOR_BGR2RGB)

print(simple.shape, simple.dtype)
print(complex.shape, complex.dtype)

print(np.unique(simple))
print(np.unique(complex))
```

```python

segments_slic = felzenszwalb(simple, scale = 100)
print(f'SLIC number of segments: {len(np.unique(segments_slic))}')

fig, ax = plt.subplots(1, 1, figsize=(10, 10), sharex=True, sharey=True)

ax.imshow(mark_boundaries(simple, segments_slic, color=(255,0,0)))



segments_slic = felzenszwalb(complex, scale=100)
print(f'SLIC number of segments: {len(np.unique(segments_slic))}')

fig, ax = plt.subplots(1, 1, figsize=(10, 10), sharex=True, sharey=True)

ax.imshow(mark_boundaries(complex, segments_slic, color=(255,0,0)))
```
